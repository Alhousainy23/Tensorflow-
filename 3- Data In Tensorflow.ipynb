## 1 - Imported Needed Libraries

```python
import tensorflow as tf      
%matplotlib inline 
import numpy as np            
import pandas as pd  
import matplotlib.pyplot as plt    
```

## 2 -  Loaded Boston Data & Show Shape Data and Display Part Of Data

```python
(X_train,y_train),(X_test,y_test) = tf.keras.datasets.boston_housing.load_data(path='boston_housing.npz',test_split=0.2,seed=113)
print('The X Train Shape is : ' , X_train.shape)
print('\t\t X Train  is : \n' , np.array(X_train[5]))
print('---------------------------------------- '*3)
print('The X Test Shape is : ' , X_test.shape)
print('\t\t X Test  is : \n' , np.array(X_test[5]))
print('---------------------------------------- '*3)
print('The y Train Shape is : ' , y_train.shape)
print('The Price Of House Number 5 Of Y Train is : ' , (y_train[5]))
print('---------------------------------------- '*3)
print('The y Test Shape is : ' , y_test.shape)
print('The Price Of House Number 5 Of Y Test is :  ' , (y_test[5]))
print('---------------------------------------- '*3)
print('The y Train Shape is : ' , y_train.shape)
print('The 25 Values Of Y Train is : ' , list(y_train[:25]))
print('---------------------------------------- '*3)
print('The y Test Shape is : ' , y_test.shape)
print('The 25 Values Of Y Test  is : ' , list(y_test[:25]))
```

## 2- cifar10 Data Using Tensorflow

```python
(x1_train,y1_train),(x1_test,y1_test)=tf.keras.datasets.cifar10.load_data()
print('The X Train Shape is : ' , x1_train.shape)
print('\t\t X Train  is : \n' , np.array(x1_train[5]))
print('---------------------------------------- '*3)
print('The X Test Shape is : ' , x1_test.shape)
print('\t\t X Test  is : \n' , np.array(x1_test[5]))
print('---------------------------------------- '*3)
print('The y Train Shape is : ' , y1_train.shape)
print('The Price Of House Number 5 Of Y Train is : ' , (y1_train[5]))
print('---------------------------------------- '*3)
print('The y Test Shape is : ' , y1_test.shape)
print('The Price Of House Number 5 Of Y Test is :  ' , (y1_test[5]))
print('---------------------------------------- '*3)
print('All y is : ' , np.unique( y1_train ))
for i in range(10):
    plt.imshow(x1_train[i])
    plt.show()
```
